```
import pandas as pd

# Let's see what we're working with here
df = pd.read_csv('data/anime_cleaned.csv')
df.columns
```

```
# Looks like there's a bunch of data that I don't need
# Let's filter the columns down to ones I care about
columns = ['title', 'title_english', 'score', 'popularity', 'rank', 'genre']
anime = df[columns]
anime
```
